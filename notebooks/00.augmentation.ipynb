```python
import os
import cv2
import matplotlib.pyplot as plt

import albumentations as A
```

```python
def show_img_and_mask(img, mask):
    plt.subplot(1,2,1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.subplot(1,2,2)
    plt.imshow(mask)
```

```python
img = cv2.imread(r"D:/WaterSegmentation/Datasets/DANU_WS_v1/test/images/6.jpg")
gray = cv2.imread(r"D:/WaterSegmentation/Datasets/DANU_WS_v1/test/masks/6.png", cv2.IMREAD_GRAYSCALE)
mask = (gray > 0) 
```

```python
show_img_and_mask(img, mask)
```

```python
transform = A.Compose([
    A.RandomBrightnessContrast(p=1),
    A.RandomShadow(p=1),
    A.RandomRotate90(p=1),
    A.RandomGridShuffle(p=1)
])
```

```python
results = transform(image=img, mask=mask)
tr_img = results["image"]
tr_mask = results["mask"]

show_img_and_mask(tr_img, tr_mask)
```
